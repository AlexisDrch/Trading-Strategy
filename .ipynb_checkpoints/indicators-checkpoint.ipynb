### Bollinger Band

```python
def get_rolling_mean(prices, n_days = 30):
    rmean = pd.Series(data = np.full(len(prices), fill_value= np.NaN), index = prices.index, name = 'rmean')
    for t in range(n_days, len(prices)):
        rmean.iloc[t] = idc.get_s_rolling_mean(prices, t, n_days)#prices[t-n_days:t].values.mean()
    return rmean

def get_rolling_stdev(prices, n_days = 30):
    stdev = pd.Series(data = np.full(len(prices), fill_value= np.NaN), index = prices.index, name = 'stdev')
    for t in range(n_days, len(prices)):
        stdev.iloc[t] = idc.get_s_rolling_stdev(prices, t, n_days) #prices[t-n_days:t].values.std()
    return stdev
        
def get_lower_upper_band(rm, rstd):
    upper_band = rm + 2*rstd
    lower_band = rm - 2*rstd
    return lower_band, upper_band 


def get_sma(prices, rmean, n_days = 30):
    sma  = pd.Series(data = np.full(len(prices), fill_value=np.NaN), index = prices.index, name = 'sma')
    for idx, val in sma[n_days:].iteritems():
        sma.loc[idx] = prices.loc[idx, 'JPM'] / rmean.loc[idx] - 1
    return sma
```

```python
def get_bbands(prices, n_days):
    rmean = get_rolling_mean(prices, n_days)
    rstd = get_rolling_stdev(prices, n_days)
    bb = pd.Series(data = np.full(len(prices), fill_value= np.NaN), index = prices.index, name='bband')
    for idx, _ in bb[n_days:].iteritems():
        bb.loc[idx] = (prices.loc[idx,'JPM'] - rmean.loc[idx]) / (2 * rstd.loc[idx]) 
    return bb
```

```python
n_days = 40
ax = in_df_jpm_prices.plot(title="BBands over JPM")
in_df_jpm_rmean = get_rolling_mean(in_df_jpm_prices, n_days)
in_df_jpm_rstd = get_rolling_stdev(in_df_jpm_prices, n_days)
in_df_jpm_sma = get_sma(in_df_jpm_prices, in_df_jpm_rmean, n_days)
in_df_jpm_low_band, in_df_jpm_up_band = get_lower_upper_band(in_df_jpm_rmean, in_df_jpm_rstd)

in_df_jpm_rmean.plot(ax=ax)
in_df_jpm_low_band.plot(ax = ax)
in_df_jpm_up_band.plot(ax = ax)
ax.legend(['JPM', 'r_mean', 'lower band', 'upper band'])
ax.set_xlabel('Date')
ax.set_ylabel('Price')
plt.show()
```

```python
in_df_jpm_bband = get_bbands(in_df_jpm_prices, n_days)

in_df_jpm_prices.tail(n=5)
in_df_jpm = pd.concat([in_df_jpm_prices, in_df_jpm_rmean, in_df_jpm_rstd, in_df_jpm_bband], axis = 1)
in_df_jpm.tail(n=5)
```

```python
in_df_buy_signals = in_df_jpm_bband[in_df_jpm_bband < -1]
in_df_sell_signals = in_df_jpm_bband[in_df_jpm_bband > 1]
#print(in_df_sell_signals)
idx = in_df_jpm_prices.index

fig = plt.figure(figsize=(18,10))
ax = fig.add_subplot(3, 1, 1, title="BBands - JPM")
ax.set_xticklabels([])
in_df_jpm_prices.plot(ax=ax)
in_df_jpm_rmean.plot(ax = ax)
in_df_jpm_low_band.plot(ax = ax)
in_df_jpm_up_band.plot(ax = ax)
ax.legend(['JPM', 'r_mean', 'lower band', 'upper band'])
ax.set_xlabel('Date')
ax.set_ylabel('Price')
plt.grid(True)

bx = fig.add_subplot(3, 1, 2)
in_df_jpm_bband.plot(ax=bx, c = 'black')
bx.set_xlabel('Date')
bx.set_ylabel('bband')
bx.set_xticklabels([])
bx.hlines(1, idx[0], idx[-1], colors='red', label='sell signal')
bx.hlines(-1, idx[0], idx[-1], colors='green', label = 'buy signal')
bx.legend(loc = 'best')
plt.grid(True)
plt.setp(plt.gca().get_xticklabels(), rotation=30)

bx = fig.add_subplot(3, 1, 3)
in_df_jpm_sma.plot(ax=bx, c = 'purple')
bx.set_xlabel('Date')
bx.set_ylabel('sma')
#bx.hlines(1, idx[0], idx[-1], colors='red', label='sell signal')
#bx.hlines(-1, idx[0], idx[-1], colors='green', label = 'buy signal')
bx.legend(loc = 'best')
plt.grid(True)
plt.setp(plt.gca().get_xticklabels(), rotation=30)
plt.show()
```

### 2. Momentum

```python
def get_momentum(prices, n_moment):
    momentum = pd.Series(data = np.full(len(prices), fill_value=np.NaN), index = prices.index, name ='momentum')
    for i in range(n_moment, len(prices)):
        momentum.iloc[i] = idc.get_s_momentum(#prices.iloc[i]['JPM'] / prices.iloc[i-n_moment]['JPM'] - 1
    return momentum
```

```python
n_moment = 20
in_df_jpm_momentum_5  = get_momentum(in_df_jpm_prices, 5)
in_df_jpm_momentum_20 = get_momentum(in_df_jpm_prices, 20)
in_df_jpm_momentum_50 = get_momentum(in_df_jpm_prices, 50)
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(2, 1, 1, title = 'JPM - momentum')
ax.set_xticklabels([])
in_df_jpm_prices.plot(ax=ax)
ax.legend(['JPM'])
ax.set_xlabel('Date')
ax.set_ylabel('Price')
plt.grid(True)

bx = fig.add_subplot(2, 1, 2)
in_df_jpm_momentum_5.plot(ax = bx, c = 'green')
in_df_jpm_momentum_20.plot(ax = bx, c = 'red')
in_df_jpm_momentum_50.plot(ax = bx, c = 'black')
bx.set_xlabel('Date')
bx.set_ylabel('momentum')
bx.legend(['5 days','20 days', '50 days'])
plt.grid(True)
plt.setp(plt.gca().get_xticklabels(), rotation=30)
plt.show()
```

### 3. Commodity Channel Index (CCI)

The commodity channel index (CCI) is an oscillator which was originally introduced by Donald Lambert in 1980. CCI can be used to identify cyclical turns across asset classes, be it commodities, indices, stocks, or ETFs. CCI is also used by traders to identify overbought/oversold levels for securities. (ref. https://www.quantinsti.com/blog/build-technical-indicators-in-python/)
